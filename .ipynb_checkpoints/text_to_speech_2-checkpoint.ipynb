```python
# This file is dedicated to name entity pronounciation task. The idea is to record all the syllables of 
# Devnagari and then play them very smoothly to mimic the pronouciation of the name entity given a true 
# spelling.
```

```python
# The way the task is achived is that
# 1. We have recorded all the Vyanjana Varnas with two different sounds. One is for the starting sound and 
# other one is with Halant mix. 
# 2. We have recorded all the Swara Varnas with properly cropped from true word pronunciations.

# Vyanjana Varnas - k kh g gh ch chh j jh t th d dh n p ph b bh m y r l w s sh - 24*2 = 48 recording (sounds).
# Swara Varnas - a aa i u e o - 6*2 = 12 , For a new accent it's 12 more sounds, that's all.
```

```python
# Now getting the sounds. 
# Importing necessary packages
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn as sk
import tensorflow as tf
import librosa
from IPython.display import Audio
import random
import soundfile as sf
import pyrubberband
```

```python
# Now we need a long audio file that will contain all the words distinctly. 

# Let's do a fine cropping program based on amplitude. 

# Importing the audio.
ar, sr = librosa.load('data/system_2/poc/ankush.wav')

print(ar.shape)
print(sr)

Audio(data=ar, rate=sr)
```

```python
# Let's import all of our recordings. 
file_list = os.listdir('data/system_2/sounds')
print(len(file_list))
print(file_list)
```

```python
sound_dict = {}

for file_name in file_list:
    inst_key = file_name.split('.')[0]
    inst_arr, inst_sr = librosa.load('data/system_2/sounds/{}'.format(file_name))
    sound_dict[inst_key] = inst_arr
    
display(sound_dict)
```

```python
# Let's play a file from the dict
sound = sound_dict['aa_e']
Audio(sound, rate=sr)
```

```python
%%time
# The  accent. 
# This is not needed with the newly accent grabbing idea. 

# Testing
# name = 'a_s n_e k_e u_s sh_e'
# name = 's_e aa_s i_e'

# Hasibul -> dictionary # Possible 
name = 'h_s a_s s_s i_m b_s u_e l_e'

# name = 'n_s e_s h_s aa_s l_e'
# name = 'r_s o_s l_s e_m k_s s_e'
# name = 'r_s o_s d_e r_s i_m g_s o_e'


# Name_1
# name = 'b_s e_s n_s u_m dh_s a_m r_s a_m n_e _ m_s u_s r_s a_m l_s i_m n_s aa_e th_e _ ch_s o_s k_e r_s o_m w_s o_m r_s t_s i_e'
# name = 'b_s e n_s u dh_s a r_s n_e'

# Name_2
# name = 'm_s aa_s h_s y_s u_m sh_e m_s a_m t_s i_e _ ch_s a_s t_s o_m p_s aa_m dh_s y_s aa_m y_s a_e _ bh_s e_s d_s e_e'
# name = 'm_s a h_s y_s u sh_s m_s a t_s i'

# Real Names
# Rowan Philip Alex
# name = 'r_s o_s w_s a_m n_e _ ph_s i_s l_s i_e p_e _ aa_s l_s e_e k_e s_e'

# Xitij Patel
# name = 'k_e s_s i_s t_s i_e j_e _ p_s a_s t_s e_e l_e'

name_ar = name.split(' ')
print(name_ar)
```

```python
%%time
# Re-assembling the words.
final_voice = np.array([])

for character in name_ar:
    inst_audio_ar = sound_dict[character]
    
    if character in ['_']:  
#         inst_audio_ar = inst_audio_ar[0 : 500]
        pass 
    
    if character in ['a', 'aa', 'i', 'u', 'e', 'o']:  
        pass    
    
    else:
        inst_audio_ar = pyrubberband.pyrb.time_stretch(inst_audio_ar, sr, 0.5, rbargs=None)
        pass
        
    final_voice = np.concatenate([final_voice, inst_audio_ar])

# Speed shifting
final_voice = pyrubberband.pyrb.time_stretch(final_voice, sr, 2, rbargs=None)

# # Pitch shifting
# final_voice = pyrubberband.pyrb.pitch_shift(final_voice, sr, 1, rbargs=None)

# # Volume shifting
# final_voice = final_voice - 0
    
# Write the final audio.
sf.write('output_{}.wav'.format(str(2)), final_voice, int(sr*1.0), 'PCM_24')

Audio(final_voice, rate=sr)
```
