```python
# Here what we are doing is that, we are taking a a word and converts it to it's character sound array.

end = '\n\n' 

# Stands for Swaravarna set.
sv_set = [
    'a', 'a2', 'i', 'u', 'e', 'o',
    'aa2', 'ai', 'au', 'ae', 'ao',
    'a2a', 'a2i', 'a2u', 'a2e', 'a2o',
    'ia', 'ia2', 'iu', 'ie', 'io',
    'ua', 'ua2', 'ui', 'ie', 'uo',
    'ea', 'ea2', 'ei', 'eu', 'eo',
    'oa', 'oa2', 'oi', 'ou', 'oe', 
]

# Stands for Vyanjanavarna set.
vv_set = [
    'k', 'kh', 'g', 'gh', 'n2',
    'ch', 'chh', 'j', 'jh',
    't', 'th', 'd', 'dh',
    't2', 'th2', 'd2', 'dh2', 'n',
    'p', 'ph', 'b', 'bh', 'm',
    'r', 'l', 'sh', 's', 'h',
    'f', 'v', 'z', 
]

# Stands for Punctuation set. 
p_set = [
    ',', '.', '?', '!'
]

char_set = sv_set + vv_set + p_set
# print(char_set)

# We will be taking 4 different lists.
char_1 = []
char_2 = []
char_3 = []
char_4 = []

for char in char_set:
    if len(char) == 4:
        char_4.append(char)
    if len(char) == 3:
        char_3.append(char)
    if len(char) == 2:
        char_2.append(char)
    if len(char) == 1:
        char_1.append(char)

# print(char_1, end=end)
# print(char_2, end=end)
# print(char_3, end=end)
# print(char_4, end=end)

# Formatter 1. This formatter converts a word to the elements of characters and returns a list.
def word_to_chars(word, char_1, char_2, char_3, char_4):
    # word = 'abhishek'
    word_to_char_ar = []

    i = 0
    final_index = len(word)
    # print(i, final_index, end=end)

    # Writing the word_to_char_formatter script.
    while i < final_index:
        # Check if the character is in the char_4.
        try:
            inst_char = word[i : i+4]
            if inst_char in char_4:
                word_to_char_ar.append(inst_char)
                i = i + 4
                continue
        except:
            pass

        # Check if the character is in the char_3.
        try:
            inst_char = word[i : i+3]
            if inst_char in char_3:
                word_to_char_ar.append(inst_char)
                i = i + 3
                continue
        except:
            pass

        # Check if the character is in the char_2.
        try:
            inst_char = word[i : i+2]
            if inst_char in char_2:
                word_to_char_ar.append(inst_char)
                i = i + 2
                continue
        except:
            pass

        # Check if the character is in the char_1.
        try:
            inst_char = word[i : i+1]
            if inst_char in char_1:
                word_to_char_ar.append(inst_char)
                i = i + 1
                continue
            else:
                # If the character was not found anywhere append '_na_' for it.
                word_to_char_ar.append('_na_')
                i = i + 1
        except:
            pass
        
    return word_to_char_ar

# Using the function.
word = 'kaese'
word_to_char_ar = word_to_chars(word, char_1, char_2, char_3, char_4)
print(word_to_char_ar)

```

```python

# Then we need a script that will only keep the Swaravarnas, spaces, ',', '.', '?', '!'  and removes all other Vyanjavarnas 
# with 'k'.
def k_replaced_sentence_maker(text):
    text_ar = text.split(' ')
    
    text_k_ar = [] 

    for i in range(len(text_ar)):
        inst_word = text_ar[i]
        inst_word_to_char_ar = word_to_chars(inst_word, char_1, char_2, char_3, char_4)

        for j in range(len(inst_word_to_char_ar)):
            inst_char = inst_word_to_char_ar[j]
            if inst_char in vv_set:
                inst_word_to_char_ar[j] = 'k'

        inst_word = ''.join(inst_word_to_char_ar)
        text_k_ar.append(inst_word)

    text_k = ' '.join(text_k_ar)
    
    return text_k

# Using the function.
text = 'di em a2e fainans me a2pka2 sua2gat hae. a2pka2 na2m ankush hae. hama2re gra2hak seva2 pratinidh2i se ba2t karne ke liye ek daba2en.'
text_k = k_replaced_sentence_maker(text)
print(text, end=end)
print(text_k)
```

```python

# Creating the char_fluctuator formatter.

# For Swaravarnas
# 1st priority.
# 1 - Before comma, the Swaravarna sound will be middle. 
# 2 - Before full stop, the Swaravarna will be having the ending sound.
# 3 - Before question mark, the Swaravarna sound will be starting.
# 4 - Before exclamation mark, the Swaravarna sound will be starting.

# 2nd priority.
# 1- For the first Swaravarna of any word it's going to be be starting.
# 2- For the last Swaravarna while sentence is on continue, the Swaravarna will be middle.
# 3- For the middle Swaravarna it's going to be middle always. 

# For Vyanjanavarnas
# 1- If there's a Swaravarna after the Vyanjavarna, then it should be starting.
# 2- Else it will be ending. 

# Now let's write the codes.
def chars_fluctuator(word_to_char_ar):
    word_to_char_fl_ar = []

    i = 0
    final_index = len(word_to_char_ar)
    first_sv = True

    while i < final_index:
        inst_char = word_to_char_ar[i]

        # Handle if it's a Vyanjavarna.
        if inst_char in vv_set:
            try:
                inst_char_after = word_to_char_ar[i+1]
                if inst_char_after in sv_set:
                    word_to_char_fl_ar.append(inst_char + '_s')
                    i = i + 1
                    continue
                if inst_char_after in vv_set:
                    word_to_char_fl_ar.append(inst_char + '_e')
                    word_to_char_fl_ar.append('_g_')
                    i = i + 1
                    continue
                if inst_char_after in p_set:
                    word_to_char_fl_ar.append(inst_char + '_e')
                    i = i + 1
                    continue
            except:
                word_to_char_fl_ar.append(inst_char + '_e')
                i = i + 1
                continue

        # Handle if it's a Swaravarna.         
        if inst_char in sv_set:
            # Applying the 1st priority.
            # Detect if it's the last Swaravarna.
            is_last = True
            for j in range(i+1, final_index):
                inst_char_next = word_to_char_ar[j]
                if inst_char_next in sv_set:
                    is_last = False
                    break

            if is_last == True:
                # See if there's any punctuations at the end.
                last_char = word_to_char_ar[-1]

                if last_char == ',':
                    word_to_char_fl_ar.append(inst_char + '_m')
                    i = i + 1
                    continue
                if last_char == '.':
                    word_to_char_fl_ar.append(inst_char + '_e')
                    i = i + 1
                    continue
                if last_char == '?':
                    word_to_char_fl_ar.append(inst_char + '_s')
                    i = i + 1
                    continue
                if last_char == '!':
                    word_to_char_fl_ar.append(inst_char + '_s')
                    i = i + 1
                    continue
                else:
                    word_to_char_fl_ar.append(inst_char + '_m')
                    i = i + 1
                    continue

            # Applying the 2nd priority.  
            if is_last == False:
                if first_sv == True:
                    word_to_char_fl_ar.append(inst_char + '_s')
                    first_sv = False
                    i = i + 1
                    continue
                else:
                    word_to_char_fl_ar.append(inst_char + '_m')
                    i = i + 1
                    continue
        
        if inst_char == '_na_':
            word_to_char_fl_ar.append('_na_')
            i = i + 1
            continue
        
        # This line is going to be executed all the time to increament i value.    
        i = i + 1
        
    return word_to_char_fl_ar

word = 'an2kush?'
word_to_char_ar = word_to_chars(word, char_1, char_2, char_3, char_4)
print(word_to_char_ar)

# Using the function.
word_to_char_fl_ar = chars_fluctuator(word_to_char_ar)
print(word_to_char_fl_ar)
        
```

```python
# Word Pronouncer
import librosa
import numpy as np
import soundfile as sf

def word_pronouncer(word_to_char_fl_ar, style, vv_volume, sv_volume, yukt_clarity=0.2):
    _g_ = np.array([0 for i in range(int(yukt_clarity*48000))])

    word_pronunciation_ar = []

    for i in range(len(word_to_char_fl_ar)):
        inst_char = word_to_char_fl_ar[i]
        inst_char_pure = inst_char.split('_')[0]

        if inst_char_pure in vv_set:
            inst_sound_ar, inst_sr = librosa.load('data/vv_set/{}'.format(inst_char + '.wav'), 48000)
            inst_sound_ar = vv_volume * inst_sound_ar
            word_pronunciation_ar = np.concatenate([word_pronunciation_ar, inst_sound_ar])
        if inst_char == '_g_':
            inst_sound_ar = _g_
            word_pronunciation_ar = np.concatenate([word_pronunciation_ar, inst_sound_ar])
        if inst_char_pure in sv_set:
            inst_sound_ar, inst_sr = librosa.load('data/sv_sets/{}/{}'.format(style, inst_char + '.wav'), 48000)
            inst_sound_ar = sv_volume * inst_sound_ar
            word_pronunciation_ar = np.concatenate([word_pronunciation_ar, inst_sound_ar])
            
    return word_pronunciation_ar

# Using the function
yukt_clarity = 0.05
style = 'style_1'
vv_volume = 0.9
sv_volume = 0.9

word_pronunciation_ar = word_pronouncer(word_to_char_fl_ar, style, vv_volume, sv_volume, yukt_clarity)
sf.write('word.wav', word_pronunciation_ar, 48000, 'PCM_24')
```

```python

# Now here we are going to write a function to pronouce a sentence. 
def sentnece_pronouncer(words, word_gap):
    words_ar = words.split(' ')

    _wg_ = np.array([0 for i in range(int(word_gap*48000))])

    final_sentence_ar = np.array([])

    for word in words_ar:
        word_to_char_ar = word_to_chars(word, char_1, char_2, char_3, char_4)
        word_to_char_fl_ar = chars_fluctuator(word_to_char_ar)
        word_pronunciation_ar = word_pronouncer(word_to_char_fl_ar, style, vv_volume, sv_volume, yukt_clarity)
        final_sentence_ar = np.concatenate([final_sentence_ar, word_pronunciation_ar, _wg_])
    
    return final_sentence_ar

# Using the function
words = 'kankan, ankush shanku. shashn2k'
word_gap = 0.5

final_sentence_ar = sentnece_pronouncer(words, word_gap)
sf.write('sentence.wav', final_sentence_ar, 48000, 'PCM_24')
    
```
